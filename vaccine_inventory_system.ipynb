```python
%%writefile vaccine_inventory.c
#include<stdio.h>
#include<string.h>
#include<stdlib.h>

#define MAX_VACCINES 5
#define USERNAME "charran"
#define PASSWORD "123"

typedef struct {
    char code[10];
    char name[30];
    char country[30];
    int dosage_required;
    int storage;
} Vaccine;

void initialize_vaccines(Vaccine vaccines[]);
void menu(Vaccine vaccines[]);
void receive_vaccine(Vaccine* vaccine);
void distribute_vaccine(Vaccine* vaccine);
void display_vaccines(Vaccine vaccines[]);
void search_vaccine(Vaccine vaccines[]);
void save_to_file(Vaccine vaccines[]);

int main() {
    char username[20], password[20];

    do {
        printf("\nEnter your username: ");
        scanf("%s", username);

        printf("Enter your password: ");
        scanf("%s", password);

        if (strcmp(username, USERNAME) == 0 && strcmp(password, PASSWORD) == 0) {
            printf("\nWelcome! Login successful!\n");
            break;
        } else {
            printf("\nInvalid credentials. Please try again.\n");
        }
    } while (1);

    Vaccine vaccines[MAX_VACCINES];
    initialize_vaccines(vaccines);

    menu(vaccines);

    return 0;
}

void initialize_vaccines(Vaccine vaccines[]) {
    strcpy(vaccines[0].code, "pf");
    strcpy(vaccines[0].name, "Pfizer");
    strcpy(vaccines[0].country, "USA");
    vaccines[0].dosage_required = 2;
    vaccines[0].storage = 1000000;

    strcpy(vaccines[1].code, "sv");
    strcpy(vaccines[1].name, "Sinovac");
    strcpy(vaccines[1].country, "China");
    vaccines[1].dosage_required = 2;
    vaccines[1].storage = 1000000;

    strcpy(vaccines[2].code, "az");
    strcpy(vaccines[2].name, "AstraZeneca");
    strcpy(vaccines[2].country, "UK");
    vaccines[2].dosage_required = 2;
    vaccines[2].storage = 1000000;

    strcpy(vaccines[3].code, "sp");
    strcpy(vaccines[3].name, "Sputnik V");
    strcpy(vaccines[3].country, "Russia");
    vaccines[3].dosage_required = 2;
    vaccines[3].storage = 1000000;

    strcpy(vaccines[4].code, "cs");
    strcpy(vaccines[4].name, "CanSinoBio");
    strcpy(vaccines[4].country, "China");
    vaccines[4].dosage_required = 1;
    vaccines[4].storage = 1000000;
}

void menu(Vaccine vaccines[]) {
    int choice;

    do {
        printf("\n=== Vaccine Inventory Management System ===\n");
        printf("1. Receive Vaccines\n");
        printf("2. Distribute Vaccines\n");
        printf("3. Display Vaccine Storage\n");
        printf("4. Search Vaccine\n");
        printf("5. Exit\n");
        printf("Enter your choice: ");
        scanf("%d", &choice);

        switch (choice) {
            case 1:
                for (int i = 0; i < MAX_VACCINES; i++) {
                    printf("%d. %s\n", i + 1, vaccines[i].name);
                }
                printf("Select vaccine to receive (1-%d): ", MAX_VACCINES);
                scanf("%d", &choice);
                if (choice >= 1 && choice <= MAX_VACCINES) {
                    receive_vaccine(&vaccines[choice - 1]);
                } else {
                    printf("Invalid choice.\n");
                }
                break;
            case 2:
                for (int i = 0; i < MAX_VACCINES; i++) {
                    printf("%d. %s\n", i + 1, vaccines[i].name);
                }
                printf("Select vaccine to distribute (1-%d): ", MAX_VACCINES);
                scanf("%d", &choice);
                if (choice >= 1 && choice <= MAX_VACCINES) {
                    distribute_vaccine(&vaccines[choice - 1]);
                } else {
                    printf("Invalid choice.\n");
                }
                break;
            case 3:
                display_vaccines(vaccines);
                break;
            case 4:
                search_vaccine(vaccines);
                break;
            case 5:
                save_to_file(vaccines);
                printf("Exiting and saving data...\n");
                return;
            default:
                printf("Invalid choice. Please try again.\n");
        }
    } while (1);
}

void receive_vaccine(Vaccine* vaccine) {
    int amount;
    printf("Enter the number of doses received for %s: ", vaccine->name);
    scanf("%d", &amount);
    vaccine->storage += amount;
    printf("Updated storage for %s: %d\n", vaccine->name, vaccine->storage);
}

void distribute_vaccine(Vaccine* vaccine) {
    int amount;
    printf("Enter the number of doses distributed for %s: ", vaccine->name);
    scanf("%d", &amount);
    if (amount > vaccine->storage) {
        printf("Error: Not enough storage to distribute.\n");
    } else {
        vaccine->storage -= amount;
        printf("Updated storage for %s: %d\n", vaccine->name, vaccine->storage);
    }
}

void display_vaccines(Vaccine vaccines[]) {
    printf("\nVaccine Code\tName\t\tCountry\t\tDosage\tStorage\n");
    printf("-------------------------------------------------------------\n");
    for (int i = 0; i < MAX_VACCINES; i++) {
        printf("%s\t\t%s\t%s\t%d\t%d\n",
               vaccines[i].code, vaccines[i].name, vaccines[i].country,
               vaccines[i].dosage_required, vaccines[i].storage);
    }
}

void search_vaccine(Vaccine vaccines[]) {
    char code[10];
    printf("Enter vaccine code to search: ");
    scanf("%s", code);

    for (int i = 0; i < MAX_VACCINES; i++) {
        if (strcmp(vaccines[i].code, code) == 0) {
            printf("\nCode: %s\nName: %s\nCountry: %s\nDosage Required: %d\nStorage: %d\n",
                   vaccines[i].code, vaccines[i].name, vaccines[i].country,
                   vaccines[i].dosage_required, vaccines[i].storage);
            return;
        }
    }
    printf("Vaccine with code '%s' not found.\n", code);
}

void save_to_file(Vaccine vaccines[]) {
    FILE* fp = fopen("vaccine.txt", "w");

    fprintf(fp, "Vaccine Code\tName\t\tCountry\t\tDosage\tStorage\n");
    fprintf(fp, "-------------------------------------------------------------\n");
    for (int i = 0; i < MAX_VACCINES; i++) {
        fprintf(fp, "%s\t\t%s\t%s\t%d\t%d\n",
                vaccines[i].code, vaccines[i].name, vaccines[i].country,
                vaccines[i].dosage_required, vaccines[i].storage);
    }

    fclose(fp);
    printf("Data saved to 'vaccine.txt'.\n");
}
```
